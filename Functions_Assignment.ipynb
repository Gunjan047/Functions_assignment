1-> What is the difference between a function and a method in Python?

 - While both functions and methods are blocks of code designed to perform specific tasks, they differ primarily in their association with objects and classes in Python's object-oriented programming (OOP) paradigm.

   Here's a breakdown of the key distinctions:

 - Functions

     - Independent: Functions are standalone code blocks not tied to any particular object or class.

     - Called by name: You call a function directly by its name.

     - General purpose: Functions are often used for tasks that operate independently of an object's state, promoting modularity and reusability, according to upGrad.

     - No implicit first argument: Functions don't automatically receive self or cls as their first argument.

     Examples: len(), print(), or a custom function to calculate a factorial.

    
- Methods

     - Associated with objects/classes: Methods are functions defined within a class and operate on instances (objects) of that class or the class itself.

     -Accessed via dot notation: You access methods using dot notation on an object or class instance, e.g., my_list.append().

     -Object-specific behavior:
     Methods define the behaviors and actions that objects of a class can perform, often interacting with and modifying the object's data, says DataCamp.

     - Implicit self or cls argument:
      Instance methods receive the instance (self) as  the first argument, and class methods receive the class (cls) as the first argument, says Real Python.

     Examples: my_list.append(), my_string.upper(), or custom methods defined within a class like a Person class having a greet() method.

In essence, all methods are technically functions, but not all functions are methods. Functions are independent, while methods are intricately linked to the objects or classes they belong to, providing them with specific behaviors and capabilities within the OOP paradigm.

4-> What is the purpose of the `return` statement in a Python function?

 - The return statement in a Python function serves two primary purposes:
Exiting the Function:

When a return statement is encountered during the execution of a function, it immediately terminates the function's execution. Any code following the return statement within that function will not be executed.

  - Returning a Value:

The return statement can optionally be followed by an expression, whose value is then sent back to the part of the code that called the function. This returned value can be assigned to a variable, used in further calculations, or passed as an argument to another function. If a function completes its execution without encountering a return statement, or if a return statement is used without an explicit value (e.g., return), the function implicitly returns the special value None.

6-> Explain the concept of generators in Python and how they are defined?

 - In Python, a generator is a special type of function that produces a sequence of values, one at a time, using the yield keyword instead of return. This allows for memory-efficient iteration, especially when dealing with large datasets or infinite sequences. Generators are defined like regular functions but use yield to produce values and pause execution, resuming from the same point when the next value is requested.

  Key Concepts:

   - yield Keyword:

 Instead of returning a single value and exiting like a regular function, a generator uses yield to produce a value and temporarily halt execution, preserving the function's state.

  - Lazy Evaluation:

 Generators compute values only when requested, making them ideal for processing large or infinite sequences without loading the entire dataset into memory.

  - Iterator Protocol:

 Generators implicitly follow the iterator protocol, allowing them to be used in for loops and other contexts that expect iterators.

- Definition:
A generator function is defined like a regular function, but it includes one or more yield statements. The yield keyword returns a value to the caller and suspends the function's execution. When the generator is called again, it resumes execution from where it left off, picking up after the last yield statement.

7-> What are the advantages of using generators over regular functions?

 - Generators in Python offer several key advantages over regular               (nongenerator) functions:

    - Lazy Evaluation:

     - A generator yields one item at a time only when requested, rather than computing an entire sequence up front.

     - This can greatly reduce upfront computation cost, especially if you may not need all the items.

    - Memory Efficiency:

       - Because they produce items one by one, generators dont need to store the entire sequence in memory.

       - You can iterate over very large (or even infinite) sequences without running out of RAM.

    - Stateful Iteration without Boilerplate:

       -  A generator automatically remembers its local state (including variable values and the current instruction pointer) between yields.

       - You dont need to manually maintain an index or stash intermediate results.

    - Composability & Pipelines:

       - Generators can be chained together (e.g., one generator feeding into another), creating clean dataprocessing pipelines.

       - This leads to code that reads like a sequence of transformations on a stream of data.

    - Infinite Sequences :

      - You can easily write generators that model infinite streams (e.g., the Fibonacci sequence, or an everincreasing counter) because they only compute each next element on demand.

  - Improved Readability:

     - For many streaming or iterative problems, generator syntax (yield) can be more concise and clearer than managing a list and returning it at the end.



8-> What is a lambda function in Python and when is it typically used?

 - In Python, a lambda function is a small, anonymous function defined using the lambda keyword. It's used for creating concise, single-expression functions, often as arguments to higher-order functions like map, filter, and reduce. They are ideal for short, throwaway operations where defining a full function with def would be overkill.

 - Key characteristics:

    - Anonymous: Lambda functions don't have names.

    - Single expression: They can only contain a single expression, which is automatically returned.

    - Concise: They allow for creating small functions in a compact syntax.

- Typical use cases:

1. Higher-order functions:

Lambda functions are commonly used as arguments to built-in functions like map, filter, and reduce.

  - map(function, iterable): Applies the function to each item in the iterable.

  - filter(function, iterable): Filters items from the iterable based on the function's return value (True or False).

  - reduce(function, iterable): Accumulatively applies the function to the items in the iterable, reducing it to a single value.

2. Temporary or throwaway functions:

  When a function is only needed for a short, specific task, a lambda function can provide a quick and convenient solution.

3. Inline functions:

  Lambda functions can be defined directly within a larger expression, making the code more concise.

9-> Explain the purpose and usage of the `map()` function in Python?

- The map() function in Python is a built-in higher-order function used to apply a specified function to each item in an iterable (such as a list, tuple, or set) and return an iterator that yields the results.

  - Purpose:
The primary purpose of map() is to efficiently transform data within an iterable without explicitly writing a for loop. It allows for concise and readable code when applying the same operation to multiple elements of a collection.

  - Usage:
The syntax for map()

  - function:

     This is the function that will be applied to each item of the iterable. It can be a built-in function, a user-defined function, or a lambda function.

  - iterable:

     This is the sequence (e.g., list, tuple, string) whose elements will be passed to the function. map() can accept multiple iterables, in which case the function must accept a corresponding number of arguments.

- Key Characteristics:

   - Returns an Iterator:

        map() returns a map object, which is an iterator. This means it generates results on demand, making it memory-efficient, especially with large datasets. To view the results, you typically convert the map object to a list, tuple, or other collection type.

  - Functional Programming:

       It aligns with functional programming principles by applying a function to data without modifying the original iterable.

  - Readability and Performance:

      map() often leads to more readable code compared to explicit loops for simple transformations and can offer performance advantages for large datasets due to its C-optimized implementation.

10-> What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 - The map(), filter(), and reduce() functions in Python are higher-order functions used for processing iterables in a functional programming style. They differ in their purpose and the type of output they produce:


- map():

 This function applies a given function to each item in an iterable and returns an iterator that yields the results. It is used for transformation, where each element of the input iterable is transformed into a new element based on the applied function. The length of the output from map() is always the same as the input iterable.

- filter():

 This function constructs an iterator from elements of an iterable for which a function returns True. It is used for selection, where elements are filtered based on a condition specified by the function. The output from filter() may have fewer elements than the input iterable, as only elements satisfying the condition are included.

- reduce():

 This function applies a rolling computation to sequential pairs of elements in an iterable, reducing the iterable to a single cumulative result. It is used for aggregation or accumulation. reduce() is part of the functools module and needs to be imported.

```python
#1-Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    even_sum =0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
```

```python
#2- Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(input_string):
    reversed_string = ""
    for char in input_string:
        reversed_string = char + reversed_string
    return reversed_string
```

```python
#3- Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    squared_numbers = []
    for num in numbers:
        squared_numbers.append(num ** 2)
    return squared_numbers
```

```python
#4-Write a Python function that checks if a given number is prime or not from 1 to 200

def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n**0.5) + 1):  # Check up to square root of n
        if n % i == 0:
            return False
    return True

#check prime numbers from 1 to 200
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number")
```

```python
#5- Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        else:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result

# Generate 10 Fibonacci numbers
fib = FibonacciIterator(10)

for num in fib:
    print(num)
```

```python
#6- Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

for power in powers_of_two(4):
    print(power)

```

```python
#7-. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # remove trailing newline characters

```

```python
#9-Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Convert to Fahrenheit using map and lambda
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

# Print the result
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
#10- Create a Python program that uses `filter()` to remove all the vowels from a given string

def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Example usage
text = "Hello World"
result = remove_vowels(text)
print("String without vowels:", result)
```
